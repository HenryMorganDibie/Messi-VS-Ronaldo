```python
#Correlation
sns.set(style="white")

# Compute the correlation matrix
corr = df.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(12, 12))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
# Select the relevant columns for Messi and Ronaldo
messi_data = df[["Messi Goals", "Messi Assists", "Messi Games"]]
ronaldo_data = df[["Ronaldo Goals", "Ronaldo Assists", "Ronaldo Games"]]

# Compute the correlation matrix for Messi and Ronaldo
messi_corr = messi_data.corr()
ronaldo_corr = ronaldo_data.corr()

# Plot the correlation matrix heatmaps for Messi and Ronaldo
fig, axs = plt.subplots(ncols=2, figsize=(10, 5))
sns.heatmap(messi_corr, annot=True, cmap="coolwarm", ax=axs[0])
axs[0].set_title("Messi")
sns.heatmap(ronaldo_corr, annot=True, cmap="coolwarm", ax=axs[1])
axs[1].set_title("Ronaldo")
plt.show()
```

```python
# Feature Engineering 
```

```python
# Create a new variable for Messi's goals per game
df["Messi Goals Per Game"] = df["Messi Goals"] / df["Messi Games"]

# Create a new variable for Ronaldo's goals per game
df["Ronaldo Goals Per Game"] = df["Ronaldo Goals"] / df["Ronaldo Games"]

# Print the first few rows of the data to check the new variables
print(df.head())
```

```python
# Hypothesis Testing
```

```python
from scipy.stats import ttest_ind

# Create a new variable for Ronaldo's goals per game
df["Ronaldo Goals Per Game"] = df["Ronaldo Goals"] / df["Ronaldo Games"]

# Remove any rows with missing values
df.dropna(inplace=True)

# Perform a two-sample t-test to compare Messi and Ronaldo's goals per game
t_stat, p_value = ttest_ind(df["Messi Goals Per Game"], df["Ronaldo Goals Per Game"], equal_var=False)

# Print the results
print("t-statistic:", t_stat)
print("p-value:", p_value)
```
